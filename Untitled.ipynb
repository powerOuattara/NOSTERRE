```python
import pandas as pd
```

```python
df=pd.read_csv('data.csv',encoding='latin1')
df.head()
```

```python
def nettoyer_dataframe(df):
    df = df.drop_duplicates()
    df = df.dropna(subset=["Description", "Quantity", "UnitPrice", "CustomerID", "InvoiceDate","Country"])
    df = df[df["Quantity"] > 0]
    df= df[df["UnitPrice"] >= 0]
    df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])
    df["revenu"] = df["Quantity"] * df["UnitPrice"]
    df.isnull().sum()
    return df
```

```python
 nettoyer_dataframe(df)
```

```python
def calculs_commerciaux_et_analyse(df):
    df=nettoyer_dataframe(df)
    import numpy as np

    if "revenu" not in df.columns:
        df["revenu"] = df["Quantity"] * df["UnitPrice"]

    revenue = df["revenu"].to_numpy()

    total = df["revenu"].sum()
    moyenne = np.mean(revenue)
    mediane = np.median(revenue)
    ecart_type = np.std(revenue)
    cv = ecart_type / moyenne if moyenne else 0
    panier_moyen = total / df["CustomerID"].nunique()

    top_produits = df.groupby("Description")["revenu"].sum().sort_values(ascending=False).head(10)
    top_clients = df.groupby("CustomerID")["revenu"].sum().sort_values(ascending=False).head(10)

    print(f" Chiffre d'affaires total : {total:.2f} FCFA")
    print(f" Panier moyen : {panier_moyen:.2f} FCFA")
    print(f" Moyenne revenu/vente : {moyenne:.2f} FCFA | Médiane : {mediane:.2f} | Écart-type : {ecart_type:.2f}")

    print("\n Top 10 produits :")
    print(top_produits)

    print("\n Top 10 clients :")
    print(top_clients)

    top1_ca = top_produits.iloc[0]
    if top1_ca > 0.6 * total:
        print("\n Le produit le plus vendu représente plus de 60% du CA.")
        print(" Risque de dépendance, pensez à diversifier vos produits.")
    else:
        print("\n Répartition des ventes produits équilibrée.")

    if mediane < 0.5 * moyenne:
        print("\n Vos revenus sont concentrés sur quelques grosses ventes.")
        print(" Fidélisez vos meilleurs clients pour stabiliser le CA.")
    else:
        print("\n Bonne répartition du revenu entre les clients.")

    if cv > 1.5:
        print("\nVentes instables : gros pics ou chutes.")
        print(" Utilisez des campagnes de promo pour lisser.")
    else:
        print("\n Bonne stabilité des ventes.")

    commandes_par_client = df.groupby("CustomerID")["InvoiceNo"].nunique()
    clients_dormants = commandes_par_client[commandes_par_client == 1].count()
    print(f"\n Clients dormants (1 seule commande) : {clients_dormants}")

    seuil_marginal = df["revenu"].sum() * 0.01
    produits_marginaux = df.groupby("Description")["revenu"].sum()
    nb_marginal = (produits_marginaux < seuil_marginal).sum()
    print(f" Produits marginaux (<1% CA) : {nb_marginal}")

    return {
        "CA_total": total,
        "Panier_moyen": panier_moyen,
        "Moyenne": moyenne,
        "Mediane": mediane,
        "Écart_type": ecart_type,
        "CV": cv,
        "Top_produits": top_produits,
        "Top_clients": top_clients,
        "Clients_dormants": clients_dormants,
        "Nb_produits_marginaux": nb_marginal
    }
```

```python
calculs_commerciaux_et_analyse(df)
```

```python
def eda(df):
    import seaborn as sns
    import matplotlib.pyplot as plt
    import pandas as pd
    df=nettoyer_dataframe(df)
    df["revenu"] = df["Quantity"] * df["UnitPrice"]
    # Vérifie que "InvoiceDate" est bien au format datetime
    if not pd.api.types.is_datetime64_any_dtype(df["InvoiceDate"]):
        df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])

    # Création de colonnes utiles pour l’analyse temporelle
    df["mois"] = df["InvoiceDate"].dt.to_period("M")
    df["jour_semaine"] = df["InvoiceDate"].dt.day_name()
    df["heure"] = df["InvoiceDate"].dt.hour

    # 1. Boxplot du revenu
    plt.figure(figsize=(10, 4))
    sns.boxplot(x=df["revenu"])
    plt.title("Distribution du revenu par commande")
    plt.xlabel("Revenu")
    plt.tight_layout()
    plt.show()

    # 2. Revenu mensuel
    revenu_mensuel = df.groupby("mois")["revenu"].sum()
    revenu_mensuel.plot(kind='bar', color='green', figsize=(10, 4))
    plt.title("Revenu par mois")
    plt.xlabel("Mois")
    plt.ylabel("Revenu (FCFA)")
    plt.tight_layout()
    plt.show()

    # 3. Revenu par jour de la semaine
    jours_ordonnes = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
    revenu_jours = df.groupby("jour_semaine")["revenu"].sum().reindex(jours_ordonnes)
    plt.figure(figsize=(8, 4))
    revenu_jours.plot(kind="bar", color="skyblue")
    plt.title("Revenu total par jour de la semaine")
    plt.xlabel("Jour")
    plt.ylabel("Revenu (FCFA)")
    plt.tight_layout()
    plt.show()

    # 4. Revenu par heure de la journée
    revenu_heures = df.groupby("heure")["revenu"].sum()
    plt.figure(figsize=(10, 4))
    revenu_heures.plot(kind="line", marker="o", color="orange")
    plt.title("Revenu par heure de la journée")
    plt.xlabel("Heure")
    plt.ylabel("Revenu (FCFA)")
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # 5. Top 10 produits par chiffre d’affaires (version horizontale)
    top_products = df.groupby("Description")["revenu"].sum().sort_values(ascending=False).head(10)
    plt.figure(figsize=(10, 5))
    sns.barplot(x=top_products.values, y=top_products.index, palette="viridis")
    plt.title("Top 10 produits par chiffre d'affaires")
    plt.xlabel("Revenu (FCFA)")
    plt.ylabel("Produit")
    plt.tight_layout()
    plt.show()

    # 6. (Optionnel) Version verticale si tu préfères
    top_products.plot(kind="bar", figsize=(10, 4), color="orange", title="Top Produits par CA")
    plt.ylabel("Revenu (FCFA)")
    plt.xlabel("Produit")
    plt.tight_layout()
    plt.show()
```

```python
eda(df)
```

```python
def alerte_stock_faible(df, seuil=10):
    df=nettoyer_dataframe(df)
    alertes = df.groupby("StockCode")["Quantity"].sum()
    alertes = alertes[alertes < seuil]
    if not alertes.empty:
        print("⚠️ Produits avec stock faible :")
        for code, qte in alertes.items():
            print(f"Produit {code} → Quantité restante : {qte}")
    else:
        print("✅ Aucun stock critique pour le moment.")
```

```python
alerte_stock_faible(df, seuil=10)
```
